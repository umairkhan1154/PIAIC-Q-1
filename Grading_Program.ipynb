```python
# enter you name 
Name = input("enter Your Username: ")
# enter your password
Password = input("enter Your Password: ")
file = open("details.txt","a")
file.write(Name)
file.write(" ")
file.write(Password)
file.write("\n")
file.close()
'''Default Subjects are Mathematics , Physics , Chemistry  and English'''
'''if you want to add more Subjects Press 1 '''
# Main Engine
check = True #BOOLEAN VARIABLE TO SET CONDITION TRUE UNTIL CORRECT USRNAME AND PASSWORD
while check:
    username = input("Please enter your username : ")
    password = input("Please enter your password : ")  
    for line in open("details.txt","r").readlines(): # Read the lines
        login_info = line.split() # Split on the space, and store the results in a list of two strings
        if username == login_info[0] and password == login_info[1]:
            print("Correct credentials!")
            check = False
            break
        else:
            print("Incorrect credentials.")
# following are our default subjects
Subjects = [['Mathematics'],['Physics'],['Chemistry'],['English']]
#creating an empty list to store marks
num = ["FIRST WEEK","SECOND WEEK ","THIRD WEEK"]
# 1st empty list is for first week , 2nd list is for second week and third is for third week
Total_marks = [[],[],[]]
    #taking the default subject marks
for i in range(0,3):
    print(f'enter for {num[i]} marks')
    Maths = int(input("Mathematics Marks :"))
    Physics = int(input("Physics Marks :"))
    Chemistry = int(input("Chemistry Marks :"))
    English = int(input("English Marks :"))
    # appending default subjects marks in the Marks list 
    Total_marks[i].append(Maths)
    Total_marks[i].append(Physics)
    Total_marks[i].append(Chemistry)
    Total_marks[i].append(English)
#Taking users choice to check if he/she wants to add more subjects than default subject
#Taking users choice to check if he/she wants to add more subjects than default subject
Choice = int(input("Press 1 if you want to add more subjects :"))
#checking user choice that if he/she Choose 1 to add more subjects or not
if Choice != 1:
    print('"enter default subject marks"')
else:
#   if user press 1 than we ask user how many subject he/she wants to enter
    Sub = int(input("Enter Number of Subjects you want to add: "))
#   looping throuhg the for the three week marks
    for i in range(0,3):
        print(f"enter for {num[i]} ")
        for j in range(Sub):
    #       taking the name of subject
            Sub_1 = input("name of subject you want to enter: ")
    #       taking sub_1 marks
            marks = int(input("Enter Sub Marks: "))
    #       appending the new subjects to the list Subjects
            Subjects.append([Sub_1])
    #       appending the New Subject marks in the list total_Marks
            Total_marks[i].append(marks)

#marking the function to calculate the subject wise percentage
def Percentage(Numbers):
    Percentage = round((Numbers/100)*(100))
    return Percentage
#checking the grades of subject
A = 0
for i in range(0,3):
    print('\n')
    print(f'{num[i]} Marks:-')
    print('\n')
    print('Subject \t\t\t Grade \t\t\t Subject_percentage')
    print('******* \t\t\t ***** \t\t\t ******************')
    for j in range(len(Total_marks[i])):
        if Total_marks[i][j] <= 100 and Total_marks[i][j] > 90:
            Grade ="A+"
            print(f'{j+1})' , end= "")
            if len(Subjects[j][0]) <= 6:
                Tab = "\t\t\t\t"
            else:
                Tab = "\t\t\t"
            print(Subjects[j][0],Tab,Grade,'\t\t\t',Percentage(Total_marks[i][j]))

        elif Total_marks[i][j] <= 90 and Total_marks[i][j] > 80:
            A +=1
            Grade ="A"
            print(f'{j+1})' , end= "")
            if len(Subjects[j][0]) <= 6:
                Tab = "\t\t\t\t"
            else:
                Tab = "\t\t\t"
            print(Subjects[j][0],Tab,Grade,'\t\t\t',Percentage(Total_marks[i][j]))

        elif Total_marks[i][j] <= 80 and Total_marks[i][j] > 65:
            Grade ="B+"
            print(f'{j+1})' , end= "")
            if len(Subjects[j][0]) <= 6:
                Tab = "\t\t\t\t"
            else:
                Tab = "\t\t\t"
            print(Subjects[j][0],Tab,Grade,'\t\t\t',Percentage(Total_marks[i][j]))

        elif Total_marks[i][j] <= 65 and Total_marks[i][j] > 50:
            Grade ="B"
            print(f'{j+1})' , end= "")
            if len(Subjects[j][0]) <= 6:
                Tab = "\t\t\t\t"
            else:
                Tab = "\t\t\t"
            print(Subjects[j][0],Tab,Grade,'\t\t\t',Percentage(Total_marks[i][j]))
        else:
            Grade = "F"
            print(f'{j+1})' , end= "")
            if len(Subjects[j][0]) <= 6:
                Tab = "\t\t\t\t"
            else:
                Tab = "\t\t\t"
            print(Subjects[j][0],Tab,Grade,'\t\t\t',Percentage(Total_marks[i][j]))
# calculating the obtained marks
obtained_marks = 0
for i in range(0,3):
    for j in range(len(Total_marks[0])):
        obtained_marks += Total_marks[i][j]
# calculating the total marks
total_marks = 0
for i in range(len(Total_marks)):
    for j in range(len(Total_marks[i])):
        total_marks += 100
print(f'you have got {obtained_marks} Marks')
# marking function to calculate the percentage
def Total(a,b):
    global Percentage
    Percentage = (a/b) * (100)
    print("Your Percentage is",round(Percentage),"%")
Total(obtained_marks,total_marks)
# checking the final result of the user
if Percentage == 100 and Percentage > 90:
    print(f"{Name} , Your final grade is A+")
elif Percentage <= 90 and Percentage > 80:
    print(f"{Name} , Your final grade is A")
elif Percentage <= 80 and Percentage > 65:
    print(f"{Name} , Your final grade is B+")
elif Percentage <= 65 and Percentage >= 50:
    print(f"{Name} , Your final grade is B")
else:
    print(f'{Name} unfortunately you are failed')
#checking out how many A grade have you Got;
print(f'You have got {A} A grades')
```
